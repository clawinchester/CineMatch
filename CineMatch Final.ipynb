```python
# import google.generativeai as genai
# Isso é como tirar a ferramenta de telefone mágico 'google.generativeai' da caixa
# e dar um apelido mais fácil para ela: 'genai'.
# Agora, sempre que quisermos usar o telefone, é só chamar 'genai'.

# --- Configurando a Chave Secreta no Colab ---
# No menu do Colab, procure por um ícone de "chave" ou "cadeado" (Geralmente na barra lateral esquerda).
# Clique nele. Vai abrir a área de Secrets.
# Clique em "+ New secret".
# No campo "Name", digite EXATAMENTE: GOOGLE_API_KEY
# No campo "Value", cole a sua chave da API que você tem.
# Marque a caixinha "Notebook access".
# Clique em "Save secret".

# Agora, no código, vamos "pedir" essa chave secreta que está guardada no cofrinho do Colab.
# O Colab guarda os secrets em um lugar chamado 'colab.userdata'.
import google.generativeai as genai
from google.colab import userdata # Precisamos importar a parte do Colab que sabe sobre os secrets

API_KEY = userdata.get('GOOGLE_API_KEY') # Pegamos o valor do secret chamado 'GOOGLE_API_KEY'

# Agora vamos "ligar" o telefone mágico usando a chave.
# Isso prepara a ferramenta para falar com o cérebro da Google.
genai.configure(api_key=API_KEY)

# E vamos escolher qual "cérebro" específico da Google queremos usar para a tarefa de filmes.
# Existem diferentes "cérebros" (modelos) para diferentes tarefas.
# 'gemini-1.5-flash-latest' é um cérebro rápido e bom para textos.
model = genai.GenerativeModel('gemini-1.5-flash-latest')

# Rode esta célula (Shift + Enter ou botão Play).
# Se tudo deu certo (instalação e chave), ele não vai mostrar erros.
# Ele apenas preparou as coisas nos bastidores.
```

```python
# Agora, a parte de falar com o usuário é igual à versão anterior!
print("Oi! Bem-vindo ao CineMatch!")
print("Estou animado para te ajudar a encontrar um filme brasileiro legal!")
print("Me diga, quais filmes estrangeiros você gosta muito?")

# Pegamos o que o usuário digitar, igual antes
filmes_que_o_usuario_gosta = input("Quais filmes estrangeiros você gosta?: ")

# --- A Mágica da Sugestão (usando a API!) ---
# Pedimos para ele ANALISAR os tipos de filmes e sugerir um SIMILAR.
pergunta_para_ia = f"Analise os gêneros, temas ou estilo dos seguintes filmes que o usuário gosta: {filmes_que_o_usuario_gosta}. Com base nessa análise, sugira APENAS O NOME de UM filme brasileiro que tenha gêneros, temas ou estilo SIMILARES. Responda SOMENTE com o nome do filme e nada mais."


print("\nPensando na melhor sugestão para você...")

# Isso manda a pergunta para a IA e espera a resposta
resposta_da_ia = model.generate_content(pergunta_para_ia)

# A resposta que vem da IA é um pouco complexa, mas o que a gente quer
# é só o texto que ela respondeu. Geralmente fica em 'resposta_da_ia.text'.
filme_sugerido = resposta_da_ia.text

# Finalmente, mostramos a sugestão que veio da IA para o usuário!
print("\nEncontrei uma ótima sugestão para você:")
print(f"Que tal você assistir o filme brasileiro: {filme_sugerido}?")
print("Espero que você goste muito!")

# ... código posterior (se houver) ...
```
